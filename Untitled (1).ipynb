```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the dataset
data = pd.read_csv('India_Road_Population_Data.csv')
```

```python
# Let's first clean the data by replacing NaN values with 0
data.fillna(0, inplace=True)
```

```python
road_data= pd.read_csv('India_Road_Population_Data.csv')
road_data
```

```python
# Check the column names in the DataFrame
print(data.columns)
```

```python
#Comparison of Road Lengths
road_columns = ['National Highways', 'State Highways', 'District Roads', 'Rural Roads', 'Urban roads', 'Project Roads']
road_data = data[road_columns].sum()
plt.figure(figsize=(10, 6))
road_data.plot(kind='bar')
plt.title('Total Road Lengths Across India')
plt.xlabel('Road Types')
plt.ylabel('Total Length (in km)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Road Density Analysis
density_columns = ['Road Density per 1000 Sq. Km - National Highways', 
                   'Road Density per 1000 Sq. Km - State Highways', 
                   'Road Density per 1000 Sq. Km - District Roads', 
                   'Road Density per 1000 Sq. Km - Rural Roads', 
                   'Road Density per 1000 Sq. Km - Urban roads', 
                   'Road Density per 1000 Sq. Km - Project Roads']

density_data = data[density_columns].mean()
plt.figure(figsize=(10, 6))
density_data.plot(kind='bar', color='skyblue')
plt.title('Average Road Density Across India')
plt.xlabel('Road Types')
plt.ylabel('Road Density (per 1000 sq. km)')
plt.show()
```

```python
# Population Distribution and Road Infrastructure
population_columns = ['Rural Pop (2011 census)', 'Urban Pop (2011 census)']
population_data = data[population_columns].sum()

plt.figure(figsize=(8, 6))
plt.bar(population_data.index, population_data.values, color=['green', 'blue'])
plt.title('Population Distribution in India')
plt.xlabel('Population Type')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
#State-wise Analysis (example: Urban Road Density vs. Rural Road Density)
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Urban Road lngth per 1000 sq km', y='Rural Road lngth per 1000 sq km', data=data, hue='Name of the States', s=100)
plt.title('Urban Road density vs. Rural Road density by State')
plt.xlabel('Urban Road lngth per 1000 sq km')
plt.ylabel('Rural Road lngth per 1000 sq km')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Histogram: Distribution of Total Road Length
fig = px.histogram(data, x='Total road Length', y='Name of the States', nbins=20, title='Distribution of Total Road Length')
fig.show()
```

```python
# Select only columns with numerical data
numerical_data = data.select_dtypes(include=[np.number])

# Correlation Analysis
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Road Infrastructure and Population')
plt.show()
```
